DROPPING UNNECESSARY COLUMNS

```python
un_useful_cols = ['id', 'Unnamed: 0.1', 'entity_type', 'entity_id', 'parent_id',
                  'domain', 'homepage_url', 'twitter_username', 'logo_url', 'logo_width', 'logo_height',
                  'name','normalized_name', 'short_description', 'description', 'overview', 'tag_list','permalink',
                  #'country_code',
                  'region','city','state_code',  #places
                  'created_by', 'created_at', 'updated_at']
```

```python
dataset.drop(un_useful_cols, axis=1,inplace=True)
```

```python
print(dataset.shape)
dataset.head()
```

```python
if dataset.duplicated().any():
  print(dataset.duplicated().sum())
dataset.drop_duplicates(inplace=True)
```

```python
print(dataset.shape)
dataset.head()
```

DROPPING COLS WITH HIGH MOST NO'OF NULL VALUES

```python
# Calculation % of null values in a column
cols_missing_percent = dataset.isnull().sum()*100/len(dataset)
print("percentage of missing values in each columns:")
cols_missing_percent
```

```python
cols_with_many_nulls = ['first_investment_at','last_investment_at','investment_rounds','invested_companies', 'ROI']
dataset.drop(cols_with_many_nulls, axis=1,inplace=True)
```

```python
dataset.isnull().sum()
```

DROPPING ROWS TO SOLVE NULL VALS

```python
vars_min_null_set = ['status', 'category_code','country_code', 'founded_at','lat','lng']
dataset.dropna(subset = vars_min_null_set ,inplace=True)
```

```python
# For founded_at, closed_at, first_funding_at, last_funding_at, first_milestone_at, last_milestone_at

date_cols = ['founded_at', 'closed_at', 'first_funding_at', 'last_funding_at', 'first_milestone_at', 'last_milestone_at']
for i in date_cols:
  dataset[i]=pd.to_datetime(dataset[i], format='%Y-%m-%d').dt.year

dataset
```

```python
dataset['category_code'].unique()
```

```python
#category_encoading
category_others_index = dataset.category_code.value_counts()[15:].index

dataset['category_code'] = dataset['category_code'].replace(to_replace=category_others_index,value='other')
# Let's check if we've more than 20 columns

print('Unique no of category_code : ',dataset['category_code'].nunique())
dataset.category_code.value_counts()
```

```python
dataset['country_code'].unique()
```

```python
country_code_others_index = dataset['country_code'].value_counts()[25:].index
dataset['country_code'] = dataset['country_code'].replace(to_replace=country_code_others_index,value='other')
# Let's check if we've more than 25 columns
print('Unique no of country_code : ',dataset['country_code'].nunique())
dataset.country_code.value_counts()
```

```python
def isClosed(row):
    if row['status'] == 'operating' or row['status'] == 'ipo':
        return 0
    else:
        return 1
dataset['is_Closed'] = dataset.apply(lambda rw: isClosed(rw),axis=1)
dataset
```

```python
closed_mask = dataset['is_Closed'] == 0

# Update 'closed_at' where 'isClosed' is 1
dataset.loc[closed_mask, 'closed_at'] = 2024
dataset
# dataset.closed_at.dropna()
```

```python
dataset.dropna(axis=0, subset=['closed_at'], inplace=True)
print(dataset.shape)
dataset.isnull().sum()
```

```python
active_days=dataset['closed_at']-dataset['founded_at']
dataset['Active_Days']=active_days
```

```python
dataset.drop(['closed_at'], axis=1,inplace=True)
```

```python
# Calculation % of null values in a column
cols_missing_percent = dataset.isnull().sum()*100/len(dataset)
print("percentage of missing values in each columns:")
cols_missing_percent
```

```python
mean_lis = ['funding_total_usd', 'relationships']

for col in mean_lis:
  dataset[col].fillna(int(dataset[col].mean()),inplace=True)
med_lis = ['funding_rounds', 'first_funding_at', 'last_funding_at']
for col in med_lis:
  dataset[col].fillna(int(dataset[col].median()),inplace=True)
mode_lis = ['first_milestone_at', 'last_milestone_at', 'milestones']
for col in mode_lis:
  dataset[col].fillna(int(dataset[col].mode()),inplace=True)
```
